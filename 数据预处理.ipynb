```python
import glob
import sys
import os
import numpy as np
```

```python
data_path = "data-20230913"

global_min = 0
global_max = 0


def read_data_map(file_name):
    with open(os.path.join(data_path, file_name), 'r') as f:
        lines = f.readlines()
        data_map = {}
        global_min = 0
        global_max = 0
        for line in lines:
            items = line.split('\t')
            name = items[0]
            values = items[1:]
            if name not in data_map:
                data_map[name] = []
            values_float = [float(value) for value in values]
            cur_min = np.min(np.array(values_float))
            cur_max = np.max(np.array(values_float))
            global_min = cur_min if global_min > cur_min else global_min
            global_max = cur_max if global_max < cur_max else global_max
            data_map[name].append(values_float)
        return data_map, global_min, global_max

# pca_data_map, global_min, global_max = read_data_map("pca")
# print(global_min, global_max)
# bph_data_map, global_min, global_max = read_data_map("bph")
# print(global_min, global_max)
```

```python
global_min = -600
global_max = 15000
global_scale = global_max - global_min
```

```python
pca_data_mean = {name: np.mean(np.vstack((values)), axis=0) for name, values in pca_data_map.items()}
```

```python
def data_normalize(data_mean_map):
    data_norm_map = {}
    for key, values in data_mean_map.items():
        values_norm = [10*(value-global_min)/global_scale for value in values]
        data_norm_map[key] = values_norm
    return data_norm_map
```

```python
def write_data_map(data_mean_map, tag):
    for key, values in data_mean_map.items():
        real_file_name = os.path.join(data_path, tag + "-" + key + ".csv")
        np.savetxt(real_file_name, np.array(values), delimiter='\n')
        print(real_file_name)
```

```python
write_data_map(pca_data_mean, 'pca')
```

```python
def preprocess_data(file_name):
    data_map, global_min, global_max = read_data_map(file_name)
    data_mean_map = {name: np.mean(np.vstack((values)), axis=0) for name, values in data_map.items()}
    data_mean_map = data_normalize(data_mean_map)
#     print(type(data_mean_map))
    write_data_map(data_mean_map, file_name)
```

```python
preprocess_data('pca')
preprocess_data('bph')
```
